# OSI model : 7 layers
    The OSI model is a valuable tool for understanding 
    the complexities of network communication

    Application Layer: 
        Provides network services to applications, 
        like email, file transfer, etc. 
        
    Presentation Layer: 
        Deals with data formatting, encryption, 
        and compression. 
        
    Session Layer: 
        Manages communication sessions between applications.     
        
    Transport Layer: 
        Ensures reliable and ordered delivery of data segments.    
        
    Network Layer: 
        Handles routing of data packets 
        between different networks.     
        
    Data Link Layer: 
        Responsible for error-free transfer of data frames 
        between two nodes on the same network. 
   
    Physical Layer: 
         Deals with the physical aspects of the network, 
         like cables, voltage levels, and data transmission rates.     


# What is PDUs?
    A PDU is a block of information that includes both

# TCP/IP model
    TCP/IP model is more practical and 
    is the foundation for how the internet operates. 
    


# DATA FLOW
    1. Application Layer:
        creates the data and passes it to the transport layer. 
    2. Transport Layer (TCP):
        The TCP layer divides the data into segments, adds header 
        information (like port numbers and sequence numbers), 
        and ensures reliable, ordered delivery. 
    3. Internet Layer (IP):
        The IP layer adds its own header (containing source and destination IP addresses) 
        to each segment, turning it into an IP datagram. 
    4. Network Access Layer (Physical and Data Link):
        This layer formats the datagram into a frame, adding a header 
        and trailer (including a checksum for error detection). 
        The physical layer then transmits the frame over the network medium. 
    
#### Receiving a Packet (Data Flow Up the Stack):

    1. Network Access Layer:
        The physical layer receives the frame and passes it to the data link layer. 
        The data link layer verifies the checksum, removes its header and trailer, 
        and sends the datagram to the internet layer. 
    2. Internet Layer:
        The IP layer examines the header, verifies the destination IP address, 
        and if the packet is fragmented, reassembles it. It then removes the IP header 
        and passes the data to the transport layer. 
    3. Transport Layer:
        The TCP layer examines its header, checks for errors, and ensures 
        the data is in the correct order. 
        It then removes its header and passes the data to the appropriate application. 
    4. Application Layer:
        The application receives the data and processes it, 
        potentially acknowledging receipt to the sending application. 

#### Key Concepts:

    Encapsulation: Adding header information at each layer on the sending end. 
    
    De-encapsulation: Removing header information at each layer on the receiving end. 
    
    Routers: Devices that operate at the network layer and forward packets based on IP addresses. 
    
    Gateways: Devices that connect different networks and may also perform routing functions. 
    
    TCP: Provides reliable, connection-oriented communication. 
    
    IP: Provides addressing and routing function

# EXTRA !! Routing Protocols
```
Routing protocols help routers communicate with each other to share information about the network's topology. Some common types include:

Distance Vector Protocols: Routers share information about the distance (usually in hops) to reach various destinations. Examples include:
RIP (Routing Information Protocol): Uses hop count as the metric, with a maximum of 15 hops.
IGRP (Interior Gateway Routing Protocol): Developed by Cisco, it uses a combination of factors like bandwidth, delay, load, and reliability.
Link State Protocols: Routers share information about the state of their links to all other routers in the network. Examples include:
OSPF (Open Shortest Path First): Uses a link-state algorithm and calculates the shortest path using Dijkstra's algorithm.
IS-IS (Intermediate System to Intermediate System): Similar to OSPF, it also uses a link-state approach.
2. Shortest Path Algorithms

Once routers have the network topology, they use algorithms to find the shortest path:

Dijkstra’s Algorithm: This algorithm finds the shortest path from a source node to all other nodes in a weighted graph. It works by evaluating the cumulative weight of paths and choosing the least costly option at each step.
Bellman-Ford Algorithm: This algorithm can also find the shortest path and is capable of handling graphs with negative weights. However, it is generally slower than Dijkstra’s algorithm.
3. Metrics

Routers evaluate different metrics to determine the best path. These metrics can include:
- Hop Count: The number of routers (hops) a packet must pass through.
- Bandwidth: The data transfer capacity of the link.
- Latency: The time it takes for data to travel from source to destination.
- Load: The amount of traffic currently being handled by the link.
- Reliability: The stability and uptime of the connection.

4. Routing Tables

Routers maintain routing tables that store the best known paths to various destinations. These tables are updated dynamically based on the information received from routing protocols.

5. Path Selection

When a router receives a packet, it checks its routing table to determine the best next hop for the packet based on the shortest path calculated through the protocols and algorithms.

Conclusion

By using a combination of routing protocols, algorithms, metrics, and routing tables, routers are able to efficiently find and maintain the shortest paths for data packets within a network. This process is crucial for optimizing network performance and ensuring reliable data transmission.
```
