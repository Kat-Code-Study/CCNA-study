# Configure CISCO devices
## What do you need? 
####  Required Equipment:
    1.Router/Switch (Cisco, HP, Juniper, etc.)
    2.Console Cable (typically RJ45 to DB9 or USB)
    3.Terminal Emulator Software (PuTTY, SecureCRT, or built-in terminal)
    4.Computer (with a serial port or USB-to-serial adapter if needed)

# CLI = Command Line Interface
# Enter privileged mode
    Router> e?
    Router> en
    Router> enable
    |
    V

# Enter global config mode  
    Router# con?
    Router# configure terminal
    Router# conf t
    |
    V
        
# Set password
    Router(config)# enable password?
    Router(config)# enable password PASSWORD
        - not encryption password, to encryption pass
            
    Router(config)# service password-encryption
    
# Set hidden password
    Router(config)# service password-encryption
    Router(config)# enable secret HIDDENPASSWORD
    |
    V
    
# Set device name
    Router(config)# hostname R1 
    |
    V

# Interface / IP
    R1(config)# interface GigabitEthernet0/0
    R1(config-if)# ip address 192.168.1.1 255.255.255.0
    R1(config-if)# no shutdown  
    R1(config-if)# exit
    
    R1(config)# line console 0
    R1(config-line)# password MyConsolePass
    R1(config-line)# login
    R1(config-line)# exit
    
    R1(config)# exit
    |
    V
    
# Show running config
    R1> enable
    R1# show running-config
        - show all, including password
        
    R1# startup-config
        - startup-config is not present
            - cause configuration not save
    |
    V
            
# Save config 
    R1# write, wr
    R1# write memory 
    R1# copy running-config startup-config
        Destination filename startup-config? enter
            Building configuration
                ok
                
    R1# startup-config
        - now it will show 
    
# EXTRA
    no service password-encryption

    back to beginning = end, disable
