# What is a Network?
    A computer network is a digital telecommunication
    network which allows Nodes to share resources.

# What is a node?
    A physical network node is an electronic 
    device that is attached to a network
    
    for example:
    modem, hub, bridge or switch
    a printer or a host computer.

# But

# SWITCH

    . Switch have many network interfaces/ports for end hosts to connect to (usually 24+).
    . Switch provide connectivity to hosts within the same LAN (Local Area Network).
    . Switch do not provide connectivity between LANs/over the Internet.

# ROUTER

     Router have fewer network interfaces than switches.
     Router are used to provide connectivity between LANS.
     Router are therefore used to send data over the Internet.

# FIREWALL

     Firewall monitor and control network traffic based on configured rules. 
     Firewall can be placed 'inside' the network, or 'outside the network! 
    . Firewall are known as 'Next-Generation Firewalls' when they include more modern and advanced filtering capabilities.
    

## Network firewalls
    are hardware devices that filter traffic between networks.
## Host-based firewalls
    are software applications that filter traffic entering and exiting a host machine, like a PC.
